```python
import numpy as np
from sklearn.metrics import pairwise_distances_argmin
import matplotlib.pyplot as plt
```

```python
x=np.array([[1,2],[3.5,4],[0.2,1.7],[4,3.8],[9,3],[7,5.5],[4.3,7.1],[3,1],[6,9]])
```

```python
def find_random_cluster(x,n_cluster,r_seed=2):
    random_state=np.random.RandomState(r_seed)
    print('random state:',random_state)
    i=random_state.permutation(x.shape[0])[:n_cluster]
    print('permutation:',i)
    center=x[i]
    print('initial center:',center)
    print('converting center:')
    
    while True:
        label=pairwise_distances_argmin(x,center)
        print('label:',label)
        new_center=np.array([x[label==i].mean(0) for i in range(n_cluster)])
        print('updated center:',new_center)
        
        if np.all(center==new_center):
            break
        center==new_center
        print('new center:',center)
        return label,center
n_cluster=2
label,center=find_random_cluster(x,n_cluster)
    
```

```python
import pandas as pd
from sklearn import datasets
from sklearn.datasets import make_blobs
x,y=datasets.make_blobs(n_samples=4000, n_features=4, random_state=42)
```

```python
df=pd.DataFrame(x,columns=["x0","x1","x2","x3"])
data=df.copy()
data["Target"]=y
data.head()
```

```python
plt.title("scatter plot of points")
plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis', s=50)
plt.grid(True)
plt.xlabel('datapoints')
plt.ylabel('')
plt.show()
```

```python
n_cluster=3
def kmeans(n_cluster,x,y,df):
    new=df.copy()
    random=np.random.RandomState(8)
    i=random.permutation(x.shape[0])[:n_cluster]
    centroid=x[i]
    print("Initial Cluster : \n",centroid)
    plt.title("Initial Centoids")
    plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis', s=50)
    plt.scatter(centroid[:,0], centroid[:,1], s=50, marker="X", c="pink", label='centroid')
    plt.xlabel('dataset')
    plt.ylabel('')
    plt.legend()
    plt.text(centroid[0][0]-2,centroid[0][1]+0.5, centroid[0], fontsize=5, color='orange')
    plt.text(centroid[1][0]-2,centroid[1][1]+0.5, centroid[1], fontsize=5, color='black')
    plt.text(centroid[2][0]-2,centroid[2][1]+0.5, centroid[2], fontsize=5, color='red')
#     plt.text(centroid[3][0]-2,centroid[3][1]+0.5, centroid[3], fontsize=5, color='green')
#     plt.text(centroid[4][0]-2,centroid[4][1]+0.5, centroid[4], fontsize=5, color='brown')
    plt.show()
    j=1
    while True:
        labels=pairwise_distances_argmin(x,centroid)
        new_center=np.array([x[labels==i].mean(0) for i in range(n_cluster)])
        print("updated center : "+str(j),new_center)
        if np.all(centroid==new_center):
            break
        centroid=new_center
        new["labels"+str(j)]=labels
        plt.title("updated center : "+str(j))
        plt.scatter(x[:,0],x[:,1],c=y,cmap='viridis', s=50)
        plt.scatter(centroid[:,0], centroid[:,1], s=50, marker="^", c="blue", label='updated center')
        plt.xlabel('datapoints')
        plt.ylabel('')
        plt.legend()
        plt.text(centroid[0][0]-2,centroid[0][1]+0.5, centroid[0], fontsize=5, color='orange')
        plt.text(centroid[1][0]-2,centroid[1][1]+0.5, centroid[1], fontsize=5, color='black')
        plt.text(centroid[2][0]-2,centroid[2][1]+0.5, centroid[2], fontsize=5, color='red')
#         plt.text(centroid[3][0]-2,centroid[3][1]+0.5, centroid[3], fontsize=5, color='green')
#         plt.text(centroid[4][0]-2,centroid[4][1]+0.5, centroid[4], fontsize=5, color='brown')
        plt.show()
        j=j+1
    return new 


kmeans(3,x,y,df)
```

```python
from sklearn.metrics import silhouette_score,accuracy_score,confusion_matrix
from sklearn.cluster import KMeans
```

```python
k=KMeans(n_clusters=3,random_state=42)
k.fit(df)
#pt=np.array([-2.901306,7.550771]).reshape(1,-1)
y_pred=k.predict(df)
#print(y_pred)
accuracy=silhouette_score(df,y_pred)
print("Silhoutte_score:",accuracy)
```

```python
kmeans=k.cluster_centers_
print("Clusters_centroid:",kmeans)
```
