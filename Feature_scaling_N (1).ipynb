# Load data from sklearn

```python
from sklearn.datasets import load_wine
import pandas as pd

wine_data=load_wine()
col=wine_data.feature_names

df=pd.DataFrame(wine_data.data,columns=col)
df
```

# Feature Scaling By Hardcode


## 1. Min_Max_Normalization

#### Formula : res=data-min/max-min

```python
import numpy as np
```

```python
def min_max_normalization(data):
    min_val=np.min(data,axis=0)
    max_val=np.max(data,axis=0)
    res=(data-min_val)/(max_val-min_val)
    return res

res=min_max_normalization(df)
result=pd.DataFrame(res,columns=col)
result
```

## Standardization

#### Formula : data-mean/std

```python
def standardization(data):
    d_mean=np.mean(data,axis=0)
    d_std=np.std(data,axis=0)
    res=(data-d_mean)/d_std
    return res

res=standardization(df)
result=pd.DataFrame(res,columns=col)
result
```

# Feature Scaling by Library

## 1. Normalization

### 1.1 MinMaxScalar

```python
from sklearn.preprocessing import MinMaxScaler
```

```python
Min_max=MinMaxScaler()
result=Min_max.fit_transform(df)
result=pd.DataFrame(result,columns=col)
result.head()
```

### 1.2 Normlizer

```python
from sklearn.preprocessing import Normalizer
```

```python
Norm=Normalizer()
res=Norm.fit_transform(df)
result=pd.DataFrame(res,columns=col)
result.head()
```

## 2. Standardization

### 2.1 StandardScaler

```python
from sklearn.preprocessing import StandardScaler
```

```python
stand=StandardScaler()
res=stand.fit_transform(df)
result=pd.DataFrame(res,columns=col)
result.head()
```

## Robust Scaling similar to standardization

```python
from sklearn.preprocessing import RobustScaler
```

```python
rob=RobustScaler()
res=rob.fit_transform(df)
result=pd.DataFrame(res,columns=col)
result.head()
```
